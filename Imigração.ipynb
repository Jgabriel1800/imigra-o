```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/imigrantes_canada.csv')
df
```

```python
df.info()
```

**Analisar as tendências de imigração do Brasil ao longo dos anos **

Colocando o nome dos países como index

```python
df.set_index('País',inplace=True)
df
```

```python
anos=list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil=df.loc['Brasil',anos]
```

```python
brasil
```

Criando um dicionário e um novo dataframe

```python
brasil_dict={'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
dados_brasil=pd.DataFrame(brasil_dict)
dados_brasil
```

Criando gráfico

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.xlabel('Anos')
plt.ylabel('Número imigrantes')
plt.title('Imigrantes do Brasil para o Canadá')
plt.show()
```

Criando figura

```python
fig, ax=plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'],lw=3)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013',fontsize=18,loc='left')
ax.set_xlabel('Anos',fontsize=14)
ax.set_ylabel('Número imigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.grid(linestyle='--')
plt.show()
```

Dois gráficos na mesma figura

```python
fig, axs=plt.subplots(1,2,figsize=(15,5))
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980-2013')
axs[0].set_xlabel('Anos')
axs[0].set_ylabel('Número imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980-2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número imigrantes')
axs[1].grid()


plt.show()
```

Analisando os 4 maiores países da America do sul

```python
fig, axs=plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n 1980-2013')

axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')


axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')


axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')


axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Anos')
  ax.set_ylabel('Número imigrantes')


ymin=0
ymax=7000
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

plt.show()
```
